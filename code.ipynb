```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
```

```python
model = load_model("lstm_model.h5")
```

```python
with open("tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

with open("max_len.pkl", "rb") as f:
    max_len = pickle.load(f)

with open("dictionary.pkl", "rb") as f:
    dictionary = pickle.load(f)
```

```python
def predict(X_list, model):
    y_ans = model.predict(X_list, verbose=0)
    y_anss = y_ans.argmax()
    return dictionary[y_anss]
```

```python
def preprocess(line, model, tokenizer, count):
    line = line.lower()
    for _ in range(count):
        num = tokenizer.texts_to_sequences([line])
        test_pad = pad_sequences(num, maxlen=max_len, padding="pre")
        word = predict(test_pad, model)
        line += " " + word

    return line   

line = "half the population was"
count = 5
returned = preprocess(line, model, tokenizer, count)
returned
```
